```python
class JednorazovaUdalost:
    def __init__(self, kdy, kolik):
        self.kdy = kdy
        self.kolik = kolik
        
    def nastane_v_den(self, aktualni_datum):
        return self.kdy == aktualni_datum
    
    def vrat_kolik(self):
        return self.kolik;

class MesicniUdalost:
    def __init__(self, den, kolik):
        self.den = den
        self.kolik = kolik
        
    def nastane_v_den(self, aktualni_datum):
        return self.den == aktualni_datum.day
    
    def vrat_kolik(self):
        return self.kolik;

# přídali jsme zapouzdření událostí
class Udalosti:
    def __init__(self, udalosti):
        self.udalosti = udalosti
    
    def vrat_zmenu_penez_v_den(self, aktualni_datum):
        zmena_penez = 0
        for udalost in self.udalosti:
            if udalost.nastane_v_den(aktualni_datum):
                zmena_penez += udalost.vrat_kolik()

        return zmena_penez

vyplata = MesicniUdalost(25, 50000)
oprava_auta = JednorazovaUdalost(datetime(2023, 5, 25).date(), -6000)

aktualni_datum = datetime(2023, 1, 1).date()
print(f"Počáteční datum: {aktualni_datum}")

udalosti = Udalosti([vyplata, oprava_auta])
stav_penez = 0

for _ in range(0, 365):
    zmena_penez = udalosti.vrat_zmenu_penez_v_den(aktualni_datum)
    if zmena_penez != 0: # Pokud došlo ke změně peněz #odpovídá to zadání?
        stav_penez += zmena_penez # zvýšíme stav peněz
        print(f"{aktualni_datum} - {stav_penez}") # zobrazíme auktuální stav?
    aktualni_datum += timedelta(days=1)
```

# Bowling
Zdroj této kapitoly: https://isbndb.com/book/9780132350884

<img src="bowling.jpg" width="600">
* Za úkol vypočítat skóre ve hře Bowling.
* Hra se skládá z deseti kol (frame). 
* Hráč má v každém kole dva pokusy na hod koulí (roll). 
* Hráč získá počet bodů podle množství shozených kuželek (pin).
* Pokud hráč schodí všech deset kuželek v prvním hodu, říká se tomu strike.
* Pokud hráč schodí všech deset kuželek dvěma hody, říká se tomu spare.

Pravidlo pro výpočet skóre je jednoduché:
* Pokud se jedná o strike, skóre je 10 plus další dva hody.
* Pokud se jedná o spare, skóre je 10 plus další hod.
* Jinak skóre jsou dva hody v kole.

## Řešení
<img src="bowling_uml.jpg" width="600">
nebo

```python
class Hra:
    def __init__(self):
        self.hody = [0 for i in range(21)]
        self.soucasny_hod = 0
    
    def hod(self, kuzelky):
        self.hody[self.soucasny_hod] = kuzelky;
        self.soucasny_hod += 1
            
    def skore(self):
        skore = 0
        index_kola = 0
        for kolo in range(10):
            if (self.je_strike(index_kola)):
                skore += 10 + self.strike_bonus(index_kola)
                index_kola += 1
            elif (self.je_spare(index_kola)):
                skore += 10 + self.spare_bonus(index_kola)
                index_kola += 2
            else:
                skore += self.dva_hody_v_kole(index_kola)
                index_kola += 2
        return skore;
    
    def je_strike(self, index_kola):
        return self.hody[index_kola] == 10
    
    def strike_bonus(self, index_kola):
        return self.hody[index_kola + 1] + self.hody[index_kola + 2]
    
    def je_spare(self, index_kola):
        return self.hody[index_kola] + self.hody[index_kola + 1] == 10
    
    def spare_bonus(self, index_kola):
        return self.hody[index_kola + 2];
    
    def dva_hody_v_kole(self, index_kola):
        return self.hody[index_kola] + self.hody[index_kola + 1]
        

def hod_nekolikrat(pocet_hodu, kuzelky):
    for i in range(pocet_hodu):
        hra.hod(kuzelky)
        
def hod_spare():
    hod_nekolikrat(2, 5)
    
def hod_strike():
    hra.hod(10)

hra = Hra()   
hod_nekolikrat(20, 0);
assert hra.skore() == 0;

hra = Hra()
hod_nekolikrat(20, 1);
assert hra.skore() == 20;

hra = Hra()
hod_spare()
hra.hod(7)
hod_nekolikrat(17, 0)
assert hra.skore() == 24;

hra = Hra()
hod_strike()
hra.hod(2);
hra.hod(3);
hod_nekolikrat(16, 0);
assert hra.skore() == 20;

hra = Hra()
hod_nekolikrat(12, 10);
assert hra.skore() == 300;
```
